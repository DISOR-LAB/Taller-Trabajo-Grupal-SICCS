```python
# ‚¨áÔ∏è 1. Install everything once
!pip install -q spacy pandas langdetect seaborn tqdm
!python -m spacy download es_core_news_sm
!python -m spacy download en_core_web_sm

# ‚¨áÔ∏è 2. Imports
import pandas as pd, spacy, matplotlib.pyplot as plt, seaborn as sns
from langdetect import detect
from tqdm.auto import tqdm

# ‚¨áÔ∏è 3. Load the data (adjust the path as needed)
fname = "/content/df_description_unique.csv"
df_raw = pd.read_csv(fname)

# Build a single text field from existing columns
df_raw["text"] = (
    df_raw["title"].fillna("") + " " + df_raw["description"].fillna("")
).str.strip()
df = df_raw.loc[df_raw["text"].str.len() > 0, ["text"]]

# ‚¨áÔ∏è 4. Load spaCy models once
nlp_es = spacy.load("es_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")

# ‚¨áÔ∏è 5. Entity extraction with language detection
records = []
for idx, txt in tqdm(df["text"].items(), total=len(df)):
    try:
        lang = detect(txt)
    except:
        continue                       # skip gibberish

    doc = (nlp_es if lang == "es" else nlp_en)(txt)
    for ent in doc.ents:
        records.append(
            {
                "row": idx,
                "lang": lang,
                "entity": ent.text,
                "label": ent.label_,
                "label_desc": spacy.explain(ent.label_),
            }
        )

df_ent = pd.DataFrame(records)
df_ent.to_excel("entities.xlsx", index=False)
print(f"‚úÖ Saved {len(df_ent):,} entities to entities.csv")

# ‚¨áÔ∏è 6. Quick visual ‚Äî top entity types
plt.figure(figsize=(8,5))
sns.countplot(
    y="label",
    data=df_ent,
    order=df_ent["label"].value_counts().index,
    palette="Dark2"
)
plt.title("Entity frequency")
plt.xlabel("# occurrences")
plt.ylabel("spaCy label")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Assuming df_org is already loaded and contains 'description' and 'author' columns

# 1. Group by description and collect authors
description_author_mapping = df_org.groupby('description')['author'].apply(list).reset_index()

# 2. Prepare data for clustering (using descriptions)
descriptions = description_author_mapping['description']

# 3. Represent descriptions numerically using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english') # Adjust stop_words if needed
X = vectorizer.fit_transform(descriptions)

# 4. Perform clustering (using K-Means)
n_clusters = 10 # Choose a suitable number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X)
description_author_mapping['cluster'] = clusters

# 5. Perform Dimensionality Reduction with PCA
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X.toarray()) # PCA works best on dense arrays

# 6. Prepare data for plotting
plot_data = []
for index, row in description_author_mapping.iterrows():
    cluster = row['cluster']
    authors = row['author']
    # Assuming all authors of a description will have the same PCA coordinates
    pca_coords = X_pca[index]
    for author in authors:
        plot_data.append({'author': author, 'cluster': cluster, 'pca1': pca_coords[0], 'pca2': pca_coords[1], 'pca3': pca_coords[2]})

df_plot = pd.DataFrame(plot_data)

print(df_author_clusters.head())

# 7. Create a 3D Scatter Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot data points, colored by cluster
scatter = ax.scatter(df_plot['pca1'], df_plot['pca2'], df_plot['pca3'], c=df_plot['cluster'], cmap='viridis', marker='o')

ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')
ax.set_title('3D Plot of Authors Clustered by Description')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_org is already loaded and contains 'description' and 'author' columns

# 1. Group by description and collect authors
description_author_mapping = df_org.groupby('description')['author'].apply(list).reset_index()

# 2. Prepare data for clustering (using descriptions)
descriptions = description_author_mapping['description']

# 3. Represent descriptions numerically using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english') # Adjust stop_words if needed
X = vectorizer.fit_transform(descriptions)

# 4. Perform clustering (using K-Means)
n_clusters = 10 # Choose a suitable number of clusters
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X)
description_author_mapping['cluster'] = clusters

# 5. Perform Dimensionality Reduction with PCA (2 components)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X.toarray()) # PCA works best on dense arrays

# 6. Prepare data for plotting
plot_data = []
for index, row in description_author_mapping.iterrows():
    cluster = row['cluster']
    authors = row['author']
    # Assuming all authors of a description will have the same PCA coordinates
    pca_coords = X_pca[index]
    for author in authors:
        plot_data.append({'author': author, 'cluster': cluster, 'pca1': pca_coords[0], 'pca2': pca_coords[1]})

df_plot = pd.DataFrame(plot_data)

# 7. Create a 2D Scatter Plot
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_plot, x='pca1', y='pca2', hue='cluster', palette='viridis', s=50)

plt.title('Cluster de usuarios')
plt.xlabel('Componente PCA 1')
plt.ylabel('Componente PCA 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Assuming df_org is already loaded and contains 'description' and 'author' columns

# 1. Group by description and collect authors (as before)
description_author_mapping = df_org.groupby('description')['author'].apply(list).reset_index()

# 2. Prepare data for clustering (using descriptions)
descriptions = description_author_mapping['description']

# 3. Represent descriptions numerically using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english') # Adjust stop_words if needed
X = vectorizer.fit_transform(descriptions)

# 4. Elbow Method
wcss = []
range_n_clusters = range(2, 11) # Consider a range of cluster numbers

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # inertia is the WCSS

plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# 5. Silhouette Score
silhouette_scores = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

plt.figure(figsize=(8, 5))
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
